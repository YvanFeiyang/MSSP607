```python
# Add 'ticker' column
df_aapl['ticker'] = 'AAPL'
df_amzn['ticker'] = 'AMZN'
df_nflx['ticker'] = 'NFLX'
df_fb['ticker']   = 'FB'
df_goog['ticker'] = 'GOOG'

# Quick sanity check
df_aapl.head()
```

```python
# Append all dataframes into a single dataframe
faanf_df = pd.concat([df_aapl, df_amzn, df_nflx, df_fb, df_goog], ignore_index=True)
faanf_df.head()
```

```python
faanf_path = '/content/drive/MyDrive/MSSP607/e.Data/Week 05 Examples/faanf.csv'
faanf_df.to_csv(faanf_path, index=False)
```

```python
# Read the combined file
df = pd.read_csv(faanf_path)
df.head()
```

```python
# Make column names lowercase
df.columns = df.columns.str.lower()
print(df.columns)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
# Sort by 'date' and 'ticker'
df_sorted = df.sort_values(['date', 'ticker']).reset_index(drop=True)
df_sorted.head()
```

```python
# Find 7 rows with the highest value for volume
df['volume'] = df['volume'].astype('int64')
top7_volume = df_sorted.nlargest(7, 'volume')
print("Top 7 rows by volume:",top7_volume)
```

```python
id_vars = ['date', 'ticker']
value_vars = [col for col in df_sorted.columns if col not in id_vars]
```

```python
#melt
df_melted = pd.melt(
    df_sorted,
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='variable',
    value_name='value'
)

df_melted.head(20)
```
