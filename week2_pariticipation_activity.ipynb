Create a variable called 'letters' that contains ['a', 'b', 'c', 'd'].

```python
letters = ['a', 'b', 'c', 'd']
print(letters)
```

What does letters[int(int('3'*2)//11)] evaluate to? Explain why in a comment.

```python
value_3b = letters[int(int('3' * 2) // 11)]
print("3b result:", value_3b)
```

Explanation:

'3' * 2 -> repeats the string '3' twice -> 33

int('33') -> converts string to integer

33 // 11 -> the result is 3

int(3) -> guarantee integer

letters[3] -> the 4th element in the list

What does letters[-1] evaluate to? Explain why in a comment.

```python
value_3c = letters[-1]
print("3c result:", value_3c)
```

Explanation:

-1 means the first element from the end, the last element in "letters" is 'd'

What does letters[:2] evaluate to? Explain why in a comment.

```python
value_3d = letters[:2]
print("3d result:", value_3d)
```

Explanation:

[:2] means[0:2], it slices from index 0 to before index 2. So in “letters” it is ['a', 'b'].
