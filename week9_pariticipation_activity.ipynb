```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import quote_plus
from pathlib import Path
```

```python
search_term = "cats"
```

```python
base_url = "https://imgur.com/search?q="
query = quote_plus(search_term)
search_url = base_url + query
print(search_url)
```

```python
r = requests.get(search_url)
print("Search status code:", r.status_code)
```

```python
soup = BeautifulSoup(r.text, "html.parser")
img_tags = soup.find_all("img")
print("Found how many tags on the page", len(img_tags))
```

```python
img_urls = []
for img in img_tags:
    src = img.get("src")
    if not src:
        continue

    if src.startswith("//"):
        src = "https:" + src
    elif src.startswith("/"):
        src = "https://imgur.com" + src

    if "i.imgur.com" in src:
        img_urls.append(src)
```

```python
# drop same ones
img_urls = list(dict.fromkeys(img_urls))
print("Unique image URLs:", len(img_urls))
```

```python
save_dir = f"/content/drive/MyDrive/MSSP607/data/Photos_{search_term}"
os.makedirs(save_dir, exist_ok=True)
```

```python
for idx, img_url in enumerate(img_urls, start=1):
    try:
        print(f"[{idx}/{len(img_urls)}] Downloading:", img_url)
        img_response = requests.get(img_url)

        if img_response.status_code == requests.codes.ok:
            # name it
            filename = f"{search_term}_{idx}.jpg"
            file_path = os.path.join(save_dir, filename)

            file = open(file_path, "wb")
            file.write(img_response.content)
            file.close()

            print("  Saved as:", file_path)
        else:
            print("  Skipped (status code:", img_response.status_code, ")")
    except Exception as e:
        print("  Failed to download", img_url, "Error:", e)
print("Dowloaded", search_term)
```
