```python
import os
import numpy as np
import pandas as pd
```

```python
file_path = '/content/drive/MyDrive/MSSP607/e.Data/parsed.csv'
df = pd.read_csv(file_path)
print(df.columns)
df.head()
```

a. Find the 95th percentile of earthquake magnitude in Japan using magType of 'mb'

```python
mask_japan_mb = (df['magType'] == 'mb') & (df['parsed_place'] == 'Japan')
japan_mb = df.loc[mask_japan_mb, 'mag']
percentile_95_japan_mb = japan_mb.quantile(0.95)
```

```python
print("95th percentile of magnitude in Japan (magType='mb'):", percentile_95_japan_mb)
```

b. Find the percentage of earthquakes in Indonesia that were coupled with tsunamis.

```python
mask_indo = df['parsed_place'] == 'Indonesia'
indo_quakes = df.loc[mask_indo]
mask_indo_tsunami = indo_quakes['tsunami'] == 1
pct_indo_tsunami = mask_indo_tsunami.mean() * 100
```

```python
print("Percentage of earthquakes in Indonesia with tsunamis: "
      f"{pct_indo_tsunami:.2f}%")
```

d. Add a column to the dataframe indicating whether or not the earthquake happened in a country or US state that is on the Ring of Fire. Use Bolivia, Chile, Ecuador, Peru, Costa Rica, Guatemala, Mexico (be sure not to select New Mexico), Japan, Philippines, Indonesia, New Zealand, Antarctica (look for the Antarctic), Canada, Fiji, Alaska, Washington, California, Russia, Taiwan, Tonga, and the Kermadec Islands).

```python
ring_keywords = [
    'Bolivia', 'Chile', 'Ecuador', 'Peru',
    'Costa Rica', 'Guatemala',
    'Japan', 'Philippines', 'Indonesia',
    'New Zealand', 'Antarctic',
    'Canada', 'Fiji',
    'Alaska', 'Washington', 'California',
    'Russia', 'Taiwan', 'Tonga',
    'Kermadec'
]

place_series = df['parsed_place'].astype(str)
ring_mask = pd.Series(False, index=df.index)
```

```python
for kw in ring_keywords:
    ring_mask = ring_mask | place_series.str.contains(kw, na=False)
mexico_mask = place_series.str.contains('Mexico', na=False) & \
              ~place_series.str.contains('New Mexico', na=False)
```

```python
df['ring_of_fire'] = ring_mask | mexico_mask
df[['parsed_place', 'ring_of_fire']].head(20)
```

e. Calculate the number of earthquakes in the Ring of Fire locations and the number outside of them

```python
ring_counts = df['ring_of_fire'].value_counts()
num_ring = ring_counts.get(True, 0)
num_non_ring = ring_counts.get(False, 0)
```

```python
print("Number of earthquakes ON the Ring of Fire:", num_ring)
print("Number of earthquakes OUTSIDE the Ring of Fire:", num_non_ring)
```

f. Find the tsunami count along the Ring of Fire.

```python
tsunami_ring_count = df.loc[df['ring_of_fire'], 'tsunami'].sum()
```

```python

print("Tsunami count along the Ring of Fire:", int(tsunami_ring_count))
```
