```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/content/drive/MyDrive/MSSP607/e.Data/InternetSales.csv'
df = pd.read_csv(file_path, encoding='latin1')
print(df.columns)
df.head()
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Year'] = df['OrderDate'].dt.year
yearly_sales = df.groupby('Year')['SalesAmount'].sum().reset_index()
print(yearly_sales)
```

```python
years = yearly_sales['Year'].values
sales = yearly_sales['SalesAmount'].values

coef = np.polyfit(years, sales, 1)
trend = np.poly1d(coef)

print("Trend function:", trend)
```

```python
last_year = years.max()
future_years = np.arange(last_year + 1, last_year + 6)
forecast_sales = trend(future_years)

forecast_df = pd.DataFrame({
    'Year': future_years,
    'ForecastSalesAmount': forecast_sales
})

forecast_df
```
