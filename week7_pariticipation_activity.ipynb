```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/content/drive/MyDrive/MSSP607/e.Data/InternetSales.csv'
df = pd.read_csv(file_path, encoding='latin1')
print(df.columns)
df.head()
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Year'] = df['OrderDate'].dt.year
yearly_sales = df.groupby('Year')['SalesAmount'].sum().reset_index()
print(yearly_sales)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(yearly_sales['Year'], yearly_sales['SalesAmount'])

plt.xlabel('Year')
plt.ylabel('Total SalesAmount')
plt.title('Yearly Internet Sales')

plt.tight_layout()
plt.show()
```
